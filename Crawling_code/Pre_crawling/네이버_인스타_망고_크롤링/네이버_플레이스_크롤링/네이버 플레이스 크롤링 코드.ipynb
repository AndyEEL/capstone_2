#### 가게 번호 크롤링

```python
adList = []
p = 1
updatedHeight = 0
numOfTrial = 0
for i in range(1, 84):
    # 스크롤 하는데 높이가 안먹혀서 END키 두번 눌러서 갱신함
    body = driver.find_element_by_css_selector('body')
    body.send_keys(Keys.END)
    body.send_keys(Keys.END)
    time.sleep(3)
    for j in range(0, 50):
        ### 데이터 크롤링 파트
        try:
            searchAdress = driver.find_element_by_xpath('//*[@id="_list_scroll_container"]/div/div/div[2]/ul/li[{}]/div[1]/a'.format(p))
            numOfTrial = 0
        except:
            p += 1
            searchAdress = driver.find_element_by_xpath('//*[@id="_list_scroll_container"]/div/div/div[2]/ul/li[{}]/div[1]/a'.format(p))
            numOfTrial += 1
            if numOfTrial == 5:
                print("아... 뻑남 ㅋ")
                break
        # Xpath 통해서 href에 있는 주소 불러오기
        adress = searchAdress.get_attribute('href')
        # 숫자만 빼기 위해서 패턴부분 제거
        adEnd = adress.find('?entry=pll')
        adNum = adress[37:adEnd]
        # 리스트에 저장
        adList.append(adNum)
        print(adNum)
        p += 1
    print('-'*10 + 'p는 현재 ' +str(p) + '임' + '-'*10)
    
```

#### 가게 리스트 저장

```python
shopNumList = pd.DataFrame(adList)
shopNumList.to_csv('shopNumList.csv', index = False)
```
