```python
import requests
import pandas as pd
import pandas as pd 
from bs4 import BeautifulSoup

### 1. 변수 설정
address = u"https://www.mangoplate.com/search/성수" 
df_list = [] #담을 변수 배열  


#url parameters 
values = {
    "keyword" : u"성수",
    'page' : 0
}

#custom header for crawling 
custom_header = {
    'referer' : 'https://www.mangoplate.com',
    'user-agent' : 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.103 Safari/537.36' 
}
```

```python

### 2. 크롤링
# 10페이지(max)까지 메인 페이지 크롤링하기 
for page in range(1,11):
    values["page"] = page #각 페이지에 대해 수행
    params = parse.urlencode(values) #url 형식에 맞게 encoding
    url = (address + "?" + params) #url 합치기
    r = requests.get(url, params = values, headers = custom_header) #컨텐츠 불러오기
    soup = BeautifulSoup(r.text, 'html.parser') # html형태로 저장 후 예쁘게 만들기
    
    #원하는 정보만 추출하여 변수 저장
    line = 0
    for li in soup.select("figcaption > div"):
        try:
            name = li.select_one('h2.title').text
            rating = li.select_one('strong.point').text
            view = li.select_one('span.view_count').text
            review_cnt = li.select_one('span.review_count').text
            
            #dataframe으로 만들땐 반드시 딕셔너리로 저장 후 삽입
            temp_row = {
                'name' : name,
                'rating' : rating,
                'view' : view,
                'review_count' : review_cnt,
            }
            df_list.append(temp_row)
            
            line += 1
            print("row "+ str(line) + " completed")
            
            #problem : line이 selector 끝에 닿으면 안멈춤. why?
            if line == 20:
                break
            
        except BaseException as ex:
            print(ex)

    print("-------page : "+ str(page) + "-------")
    

#soup = soup.prettify() #보기 예쁘게 만들어주쎄욘 
#print(soup)
    
```
